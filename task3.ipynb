```python
sm.corr()
sns.heatmap(sm.corr(),annot=True)
plt.show()
```

## shipping analysis

```python
plt.figure(figsize=(6,6))
plt.title('ship modes')
plt.pie(sm['Ship Mode'].value_counts(),labels=sm['Ship Mode'].value_counts().index,autopct='%1.1f%%')
plt.show()
sns.countplot(x=sm['Ship Mode'])
```

## segment analysis

```python
sns.pairplot(sm,hue="Segment")
```

## category  wise analysis

```python
plt.figure(figsize=(6,6))
plt.title('Category')
plt.pie(sm['Category'].value_counts(),labels=sm['Category'].value_counts().index,autopct='%1.1f%%')
plt.show()
sns.countplot(x='Category',data=sm,palette='tab10')
```

```python
plt.figure(figsize=(8,8))
plt.title('Sub-Category',fontsize=20)
plt.pie(sm['Sub-Category'].value_counts(),labels=sm['Sub-Category'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

##  state wise profits

```python
plt.figure(figsize=(16,16))
stpr=sm.groupby(['State'])['Profit'].sum().nlargest(50)
stpr.plot.barh()
```

```python
plt.figure(figsize=(6,6))
plt.title('Region wise profit',fontsize=20)
plt.pie(sm['Region'].value_counts(),labels=sm['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

## profit vs discount

```python
plt.style.use('seaborn')
sm.plot(kind="scatter",figsize=(10,6),x='Sales',y='Profit',c='Discount',s=20,fontsize=12,colormap='plasma')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profits',data=sm)
plt.legend()
plt.show()
```

## state wise profit/loss and sales

```python
pls=sm.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
pls[:].plot.bar(color=['blue','red'],figsize=(20,20))
plt.title('prfit/loss and sales of states')
plt.xlabel('States')
plt.ylabel('prfit/loss and sales')
plt.show()
```
